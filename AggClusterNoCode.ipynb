```python
# Basic Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Geneate data with two clusters

```python
# Import make_blobs
from sklearn.datasets import make_blobs
```

```python
# Make the data with two features and 3 clusters
x,y=make_blobs(n_samples=100, n_features=2, centers=3)
```

```python
# Visualize the data
plt.scatter(x[:,0], x[:,1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Agglomerative Clustering

### Building the dendrogram
> A dendrogram, or tree-like diagram, visually represents the hierarchical relationships between data points.

 > In hierarchical clustering, each data point starts as its own cluster, and then the closest clusters are merged until all data points are in a single cluster. The dendrogram shows the order in which these merges occur and the distances between the merged clusters

```python
# Imports dendrogram and linkage from scipy.cluster.hierarchy
from scipy.cluster.hierarchy import dendrogram, linkage

# Perform hierarchical clustering using the 'ward' method (minimizes variance)
# Apply linkage on the data
z=linkage(x, method='ward')
# Plot the dendrogram
dendrogram(z)
plt.show()
```

## Using the Dendrogram to Decide the Number of Clusters:
1. Plot the dendrogram.
2. Identify the longest vertical line without any horizontal lines cutting through it. This gap represents the largest jump in distance when merging clusters, indicating that the clusters are distinct.
3. Draw a horizontal line across the dendrogram just below this longest vertical line. The number of vertical lines this horizontal line crosses
corresponds to the number of clusters you should select.

## Agglomerative Clustering

```python
# Import
from sklearn.cluster import AgglomerativeClustering
```

```python
# Apply Agglomerative Clustering
# Instantiate
agg_cluster=AgglomerativeClustering(n_clusters=3)
# fit predict
clusters=agg_cluster.fit_predict(x)
```

```python
clusters
```

```python
# Plot the data with clusters
plt.scatter(x[:,0], x[:,1],c=clusters)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

```

```python
# Silhouette Score
from sklearn.metrics import silhouette_score

silhouette_score(x, clusters)
```
