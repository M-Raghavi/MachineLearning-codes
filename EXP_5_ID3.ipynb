```python
# Create the above table using pandas
# (Create a dictionary of lists)
data={
    'A1':['True','True','False','False','False','True','True','True','False','False'],
    'A2':['Hot','Hot','Hot','Cool','Cool','Cool','Hot','Hot','Cool','Cool'],
    'A3':['High','High','High','Normal','Normal','High','High','Normal','Normal','High'],
    'Target':['No','No','Yes','Yes','Yes','No','No','Yes','Yes','Yes']

}

# Make the dataframe
df=pd.DataFrame(data)


# Check the dataframe
df
```

```python
# Import label encoder
from sklearn.preprocessing import LabelEncoder

# Instantiate the encoder
le=LabelEncoder()

# Apply label encoder to the data and store it in new variable
df_encoded=df.apply(le.fit_transform)
df_encoded
```

```python
# Split the features and target
X=df_encoded.drop('Target',axis=1)
y=df_encoded['Target']
```

```python
# Import the decision tree classifer
from sklearn.tree import DecisionTreeClassifier

# Instantiate the classifer
model=DecisionTreeClassifier(criterion='entropy')

# Fit the model
model.fit(X,y)
```

```python
# Import the decision tree plotter
from sklearn.tree import plot_tree
# Plot the tree
plt.figure(figsize=(25,10))
plot_tree(model,feature_names=df_encoded.columns,class_names=le.classes_,filled=True)
plt.show()
```

```python
#import iris dataset from sklearn
from sklearn.datasets import load_iris

iris=load_iris()

X=iris['data']
y=iris['target']

df=pd.DataFrame(X,columns=iris.feature_names)
```

```python
def mapap(X):
  return iris.target_names[X]
```

```python
df['target']=y_n
```
