# **D**ensity-**B**ased **S**patial **C**lustering of **A**pplications with **N**oise    

## Basic Imports

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline




```

## Why DBSCAN when we have *k-Means* and *Hierarchical* clustering?

```python
# Import make_moons
from sklearn.datasets import make_moons


# Generate data
X,y=make_moons(noise=0.1,random_state=42,n_samples=500)

# Visualize the data
plt.figure(figsize=(15,6))
plt.scatter(X[:,0],X[:,1])
plt.show()

```

```python
# Import KMeans
from sklearn.cluster import KMeans


# Instantiate
kmeans=KMeans(n_clusters=2)

# Fit
kmeans_clusters=kmeans.fit_predict(X)


# Visualize
plt.figure(figsize=(15,6))
plt.scatter(X[:,0],X[:,1],c=kmeans_clusters)
plt.show()


```

## DBSCAN

```python
# Import DBSCAN
from sklearn.cluster import DBSCAN


# Instantiate
dbscan=DBSCAN(eps=0.2,min_samples=10)

# Fit
dbscan_clusters=dbscan.fit_predict(X)

# Visualize
plt.figure(figsize=(12,5))
plt.scatter(X[:,0],X[:,1],c=dbscan_clusters)
plt.show()


```

```python
# 1. Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
X,y=make_blobs(random_state=42,n_samples=100)
plt.figure(figsize=(12,5))
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=1.5,min_samples=10)
dbscan_clusters=dbscan.fit_predict(X)
plt.figure(figsize=(12,5))
plt.scatter(X[:,0],X[:,1],c=dbscan_clusters)
plt.show()
```
