### `Task-4`: Build confusion matrix from scratch
<details>
<summary>Hint</summary>
<b>tn = sum((actual == 0) & (predicted == 0))</b>
</details>

```python
# Confusion matrix function
def metrics(y_true,y_pred):
  """
  Computes tp, fn, fp, & tn of given actuals and predictions.
  Args:
    y_true - Actual values
    y_pred - Predicted values
  Returns:
    tp, fn, fp, tn
  """
  tn=np.sum((y_true==0)&(y_pred==0))
  fn=np.sum((y_true==1)&(y_pred==0))
  fp=np.sum((y_true==0)&(y_pred==1))
  tp=np.sum((y_true==1)&(y_pred==1))
  return np.array([[tn,fp],[fn,tp]])
```

```python
metrics(y,y_predicted)
```

### `Task-7`: Calculate accuracy, precision, recall, and f1-score manually
<details>
<summary>Hint</summary>
<b>accuracy = (tp+tn)/(tp + fn + fp + tn)</b>
</details>

```python
# Metrics function
def compute_metrics(tp,fn,fp,tn):
  """
  Computes accuracy, precision, recall & f-1 score.
  Args:
    tp - True Positives
    fn - False Negatives
    fp - False Positives
    tn - True Negatives
  Returns:
    accuracy, precision, recall, f-1 score
  """
  accuracy=(tp+tn)/(tp+fn+fp+tn)
  precision=(tp)/(tp+fp)
  recall=tp/(tp+fn)
  f1_score=(2*precision*recall)/(precision+recall)
  return accuracy,precision,recall,f1_score
tn,fp,fn,tp=metrics(y,y_predicted).ravel()
compute_metrics(tp,fn,fp,tn)
```

```python
# Compute the prediction probabilities
y_prediction_probabilities=model.predict_proba(X)
y_prediction_probabilities
```

### `Task-10`: Compute TPR and FPR and plot the ROC

```python
# Imports
from sklearn.metrics import roc_curve,roc_auc_score
# Compute TPR and FPR
fpr, tpr, thresholds = roc_curve(y, y_prediction_probabilities_positives)
# Area under curve
plt.plot(fpr,tpr)
plt.fill_between(fpr, tpr, alpha=0.3, color='b')
plt.plot([0,1],[0,1],"r--")
plt.xlabel("TPR")
plt.ylabel("FPR")
plt.show()
```
